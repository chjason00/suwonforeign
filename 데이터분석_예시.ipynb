```python
df_birth = pd.read_csv('birth.csv',encoding="cp949")
df_grdp = pd.read_csv('GRDP.csv',encoding="cp949")
df_density = pd.read_csv('density.csv',encoding="cp949")
```

```python
# 필요한 컬럼만 뽑기 && 헤더 행과 "전국" 지수 지워주기

df_birth = df_birth[['시군구별', '2023.1']]
df_birth = df_birth.drop(0)
df_birth = df_birth.drop(1)

df_grdp = df_grdp[['시도별', '2023 p)']]
df_grdp = df_grdp.drop(0)
df_grdp = df_grdp.drop(1)

df_density = df_density[['행정구역별', '2023']]
df_density = df_density.drop(0)
```

```python
#  컬럼 이름 통일

df_birth.columns = ['지역', '합계출산율']
df_grdp.columns = ['지역', '1인당GRDP']
df_density.columns = ['지역', '인구밀도']
```

```python
# 숫자로 변환 (astype) -> KOSIS 에서 추출한 csv가 숫자값 또한 문자로 저장되어 다운되기 때문에 변환 과정을 거침

df_birth['합계출산율'] = df_birth['합계출산율'].astype(float)
df_grdp['1인당GRDP'] = df_grdp['1인당GRDP'].astype(float)
df_density['인구밀도'] = df_density['인구밀도'].astype(float)
```

```python
colors = plt.cm.tab20.colors   # 20가지 확실히 구분되는 색 팔레트
```

```python
# 합계 출산율 도넛형
plt.pie(df_birth['합계출산율'], autopct="%.1f%%", startangle=90, colors=colors)
plt.title("지역별 합계출산율")
plt.legend(df_birth['지역'], title="지역", loc="center left", bbox_to_anchor=(1, 0.5))
plt.show()
```

```python
def autopct_func(pct):
    return ('%.1f%%' % pct) if pct > 3 else ''  # 3% 이하면 공백 처리하는 함수

plt.pie(df_density['인구밀도'], autopct=autopct_func, startangle=90, colors=colors)
plt.title("지역별 인구밀도")
plt.legend(df_density['지역'], title="지역", loc="center left", bbox_to_anchor=(1, 0.5))
plt.show()
```

```python
plt.pie(df_grdp['1인당GRDP'], autopct="%.1f%%", startangle=90, pctdistance=1.2, colors=colors)   # 퍼센트를 바깥쪽으로

plt.legend(df_grdp['지역'], title="지역", loc="center left", bbox_to_anchor=(1,0.5))
plt.title("지역별 1인당 GRDP")
plt.show()
```

```python
# 세 지표를 한 번에 묶은 막대 그래프 (GRDP 만 단위, 인구밀도 천 단위 -> 출산율의 단위가 작기 때문)
df_all = df_birth.merge(df_grdp, on="지역").merge(df_density, on="지역")


# 단위 변환
df_all['1인당GRDP(만원)'] = df_all['1인당GRDP'] / 10000   # 만 단위
df_all['인구밀도(천명/km²)'] = df_all['인구밀도'] / 1000   # 천 단위


# 그래프 그리기
chart = df_all.plot(x="지역", y=["합계출산율", "인구밀도(천명/km²)", "1인당GRDP(만원)"], kind="bar", figsize=(14,6))


# 각 컬럼 평균값 계산
mean_birth = df_all["합계출산율"].mean()
mean_density = df_all["인구밀도(천명/km²)"].mean()
mean_grdp = df_all["1인당GRDP(만원)"].mean()


# 평균선 추가 (각기 다른 색/스타일)
chart.axhline(mean_birth, color="blue", linestyle="--", linewidth=1, label=f"출산율 평균 {mean_birth:.2f}")
chart.axhline(mean_density, color="orange", linestyle="--", linewidth=1, label=f"인구밀도 평균 {mean_density:.2f}")
chart.axhline(mean_grdp, color="green", linestyle="--", linewidth=1, label=f"GRDP 평균 {mean_grdp:.2f}")


plt.title("지역별 출산율·인구밀도(천 단위)·1인당 GRDP(만 단위)")
plt.xticks(rotation=90)
plt.legend()
plt.show()
```
